```python
y = dataSariMax[target]
X = dataSariMax[features]

scaler_X = MinMaxScaler()
X_scaled = scaler_X.fit_transform(X)

split = int(len(y) * 0.8)
y_train, y_test = y[:split], y[split:]
X_train, X_test = X_scaled[:split], X_scaled[split:]
```

```python
model = SARIMAX(endog=y_train, exog=X_train, order=(5,1,0))
model_fit = model.fit(disp=False)

forecast = model_fit.forecast(steps=len(y_test), exog=X_test)

rmse = np.sqrt(mean_squared_error(y_test, forecast))
r2 = r2_score(y_test, forecast)
mae = mean_absolute_error(y_test, forecast)

print(f"RMSE: {rmse:.2f}")
print(f"R²: {r2:.2f}")
print(f"MAE: {mae:.2f}")
```

```python

y_train_pred = model_fit.fittedvalues

rmse_train = np.sqrt(mean_squared_error(y_train[1:], y_train_pred[1:]))  # skip indeks pertama karena differencing
mae_train = mean_absolute_error(y_train[1:], y_train_pred[1:])
r2_train = r2_score(y_train[1:], y_train_pred[1:])

print("[TRAIN]")
print(f"RMSE: {rmse_train:.2f}")
print(f"MAE : {mae_train:.2f}")
print(f"R²  : {r2_train:.2f}")

forecast = model_fit.forecast(steps=len(y_test), exog=X_test)

rmse_test = np.sqrt(mean_squared_error(y_test, forecast))
mae_test = mean_absolute_error(y_test, forecast)
r2_test = r2_score(y_test, forecast)

print("\n[TEST]")
print(f"RMSE: {rmse_test:.2f}")
print(f"MAE : {mae_test:.2f}")
print(f"R²  : {r2_test:.2f}")
```

```python
from datetime import timedelta

forecast_days = 4
last_date = dataSariMax.index.max()  

# Buat DataFrame fitur exog untuk 4 hari ke depan
future_dates = [last_date + timedelta(days=i) for i in range(1, forecast_days + 1)]

future_features = []
for d in future_dates:
    bulan = d.month
    hari = d.day
    hari_dlm_minggu = d.weekday()
    critical = 0
    
    mean_features = dataSariMax[features].mean()
    pm10 = mean_features['pm10']
    pm25 = mean_features['pm25']
    so2 = mean_features['so2']
    co = mean_features['co']
    o3 = mean_features['o3']
    no2 = mean_features['no2']

    future_features.append([pm10, pm25, so2, co, o3, no2, bulan, hari, hari_dlm_minggu, critical])
```

```python
future_features_df = pd.DataFrame(future_features, columns=features, index=future_dates)
future_features_scaled = scaler_X.transform(future_features_df)
forecast = model_fit.forecast(steps=forecast_days, exog=future_features_scaled)

df_forecast = pd.DataFrame({
    'tanggal': future_dates,
    'prediksi_max': forecast
})

print(df_forecast)
```

```python
avg_per_category = dataSariMax.groupby('kategori')['max'].mean().sort_values()

avg_per_category = avg_per_category.dropna()
avg_per_category = avg_per_category.drop("TIDAK ADA DATA", errors='ignore')

print("Rata-rata per kategori:")
print(avg_per_category)

categories = avg_per_category.index.tolist()
means = avg_per_category.values.tolist()

# Ambil nilai tengah antar kategori sebagai batas
thresholds = [(means[i] + means[i+1]) / 2 for i in range(len(means) - 1)]
print("Thresholds:", thresholds)

def klasifikasi_dari_rata2(nilai):
    if nilai <= thresholds[0]:
        return categories[0]
    elif nilai <= thresholds[1]:
        return categories[1]
    elif len(thresholds) > 2 and nilai <= thresholds[2]:
        return categories[2]
    else:
        return categories[-1]


df_forecast['kategori'] = df_forecast['prediksi_max'].apply(klasifikasi_dari_rata2)

print("\nHasil prediksi dan klasifikasi:")
print(df_forecast)
```

```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt

df = pd.read_csv('output_diisi_tuntas.csv')

df['tanggal'] = pd.to_datetime(df['tanggal'])
df = df.sort_values('tanggal')
df.set_index('tanggal', inplace=True)

pm25 = df['pm25'].asfreq('D').dropna()

model = SARIMAX(pm25, order=(1,1,1), seasonal_order=(1,1,1,7))
result = model.fit(disp=False)

forecast = result.get_forecast(steps=4)
forecast_mean = forecast.predicted_mean
forecast_ci = forecast.conf_int()

last_actual_point = pm25.iloc[-1:]
forecast_connected = pd.concat([last_actual_point, forecast_mean])

print("Hasil Forecasting pm25 untuk 4 Hari ke Depan:")
print(forecast_mean)

plt.figure(figsize=(10,5))
plt.plot(pm25[-30:], label='Data Aktual')
plt.plot(forecast_connected, color='red', label='Prediksi 4 Hari')
plt.fill_between(forecast_ci.index, forecast_ci.iloc[:, 0], forecast_ci.iloc[:, 1], color='pink', alpha=0.3)
plt.title('Prediksi pm25 4 Hari ke Depan (SARIMAX tanpa exogenous)')
plt.legend()
plt.grid(True)
plt.show()
```
